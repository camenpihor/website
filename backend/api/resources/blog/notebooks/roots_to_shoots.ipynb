## Potential Solutions

### Cohesion Theory

[[1]](https://babel.hathitrust.org/cgi/pt?id=uva.x001659730&view=1up&seq=609)
[[2]](https://onlinelibrary.wiley.com/doi/abs/10.1034/j.1399-3054.2002.1140301.x)
[[3]](https://academic.oup.com/jxb/article/48/10/1753/510047)

### Explanation

Transpiration (see [here](#Transpiration)) sets up a tension that is transmitted to the xylem (mainly from the leaves) and then down to the roots by a continuous columns of water via the cohesive properties of water. As a water molecule is transpired (ie. evaporated from the leaf) it pulls on all the other water molecules to fill its place, and those recursively pull up all the water molecules below them through the xylem and the roots.

### Drawbacks

- requires at least 3 MPa of pressure to lift water from the roots to the foliage of a 100 meter tree
- this theory was created in the late 1800s and the techniques to measure pressure were essentially putting it in a pressure chamber and cutting the plant open which broke the closed system. Therefore, they could only measure the pressure within the xylem of pressure-chamber-sized plants indirectly, and therefore their measuresment weren't the most precise, leading them to prove their theory with faulty data
- some are thinking that the xylem walls are actually quite hydrophobic (due to lignin and infiltration of solvents through the leaves), and therefore the overall cohesiveness of the system is lower. If the cohesiveness is lower, then the amount of pressure it can withstand is also lower, leading to cavitation (bubbles forming) which break the continuous flow of water which the theory leans upon
- the exact pressures required are found in small plants, but as the size of the tree exceeds 10 meters, additional theories are needed
- there is increasing evidence that tall plants are cavitated at times of peak transpiration
    - in a cavity the pressure is positive and subatmospheric

## Gibbs–Marangoni Streaming Hypothesis

[[1]](https://onlinelibrary.wiley.com/doi/abs/10.1034/j.1399-3054.2002.1140301.x)
[[2]](https://en.wikipedia.org/wiki/Marangoni_effect)
[[3]](https://royalsocietypublishing.org/doi/10.1098/rstb.1993.0087)

### Explanation

Since a liquid with a high surface tension pulls more strongly on the surrounding liquid than one with a low surface tension, then the presence of a gradient in surface tension across a liquid-liquid or liquid-gas interface will cause the liquid to flow (move) towards the areas of high surface tension. Surface tension gradients can be caused by the concentration of the liquid or its temperature being variable (as well as some other things). 

**Side note**: water has a very high surface tension, so even variations of a few percent can cause very fast Marangoni flow. [This experiment using water, soap, and pepper flakes](https://youtu.be/Ax38VQCRsyk?t=88) (water because of its high surface tension; soap because of its low surface tension; and pepper flakes for visibility) displays this affect beautifully.

We can extend this to trees by saying that whenever their is a cavitation (air bubble) that that would actually cause water to move up the air bubble towards an area of higher tension (the leaves due to transpiration). This would only be the case in which a vessel in the xylem is only partially cavitated, in the case of full cavitation, there would be no water movement.
